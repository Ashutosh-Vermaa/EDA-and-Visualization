```python
import pandas as pd
import numpy as np
import plotly.offline as py
import plotly.graph_objects as go
pd.set_option('display.max_columns', None)
```

```python
null_val=dict(data.isnull().mean())
remove_col=[key for key in null_val.keys() if null_val[key]>0.85]
```

###### Automatically generating all the traces

```python
trend=data.groupby(['country_txt','year'])['eventid'].count().reset_index().sort_values(by= 'eventid', ascending=False)[:100]
```

```python
#etracting top 10 terrorism hit countries
trend= trend[trend.country_txt.isin(list(top10))]
trend
```

```python
#pivoting is required to create automatic traces for line chart of each country
temp=trend.pivot_table(values='eventid', index=trend.year, columns='country_txt', aggfunc='first').reset_index()
temp.head()
```

```python
df=temp.T.reset_index()
df.columns= df.iloc[0]
df=df[df.year!='year']
df.head()
```

```python
df.set_index('year', inplace=True)
df.head(2)
```

```python
#creating the traces for each country automatically
traces=[go.Scatter(
    x=df.columns,
    y=df.loc[rowname],
    mode='lines+markers',
    name=rowname
)for rowname in df.index]
```

```python
layout=go.Layout(title="attacks over years in top10 affected countries",
                 )

fig=go.Figure(data=traces, layout=layout)

py.iplot(fig)
```

### % of Unsuccessful attacks country wise


```python
country= data.groupby(['country_txt', 'success'])['eventid'].count().reset_index()
country
```

```python
pivot=country.pivot_table(columns='success', values='eventid', index='country_txt')
pivot.head(2)
```

```python
pivot.rename(columns={
    1:'successful',
    0:'unsuccessful'
}, inplace=True)
```

```python

pivot['total']= pivot.successful+pivot.unsuccessful
pivot['ratio']= pivot.unsuccessful/pivot.total
pivot.head(2)
```

```python
#stack bar plot
pivot=pivot.sort_values(by= 'total', ascending=False)[:10]
trace0= go.Bar(
    x= pivot.index,
    y= pivot.unsuccessful,
    name='Unsuccessful'

)

trace1= go.Bar(
    x= pivot.index,
    y= pivot.successful,
    name='successful'

)

chart=[trace0, trace1]

layout=go.Layout(title="Country wise Successful Vs Unsuccessful Attacks",
                 barmode='stack')

figure= go.Figure(data=chart, layout=layout)

py.iplot(figure)
```

```python
# Ratio of unsuccessful attacks for countries with more than 1000 attacks
pivot= pivot[pivot.total>=1000].sort_values(by='ratio', ascending=0)
pivot
```

```python
chart= go.Bar(
    x= pivot.index,
    y= pivot.ratio,
    marker= dict(color='rgb(220, 100, 0)')

)

layout=go.Layout(title="Country wise Unsuccessful attacks ratio"
                )

figure= go.Figure(data=chart, layout=layout)

py.iplot(figure)
```

There is not much difference between the failure percentage of attacks between the nations. Therefore, the failure of the attacks might NOT have been primarily because of the alertness of the nation

### Success rate of attacks over years


```python
temp=data.groupby(['success', 'year'])['eventid'].count().reset_index()
rate=temp.pivot_table(columns='success', values='eventid', index='year')

rate.rename(columns={
    1:'successful',
    0:'unsuccessful'
}, inplace=True)
rate['total']= rate.successful+rate.unsuccessful
rate['ratio']= rate.successful/rate.total
rate.head(2)
```

```python
chart=go.Scatter(
    x=rate.index,
    y=rate.ratio,
    mode='lines+markers'
)

layout=go.Layout(
    title="Success Rate of Attacks over years",
    xaxis= dict(title="year"),
    yaxis= dict(title="Success percentage")
)

figure=go.Figure(data=chart, layout=layout)

py.iplot(figure)
```

The success rate of attacks is declining from 2007 onwards which can be attributed to the use of technology and alterness of the intelligebce agencies around the globe, including collaboration b/w countries on in information sharing

### Most Active groups in regions


```python
reg=data.groupby([ 'region_txt','group']).agg({'eventid':'count'}).sort_values(by=['region_txt', 'eventid'], ascending=[1,0]).reset_index()
reg
```

```python
reg=reg[reg.group.isin(["Unknown"])==False].reset_index()
reg_ind=[]
visited=[]
for ind, name in enumerate(reg.region_txt):
  if name not in visited and name!="":
    reg_ind.append(ind)
    visited.append(name)


most_active=reg.loc[reg_ind]
most_active.head()
```
